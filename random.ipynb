```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.__version__)
```

```python
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

x,y
```

```python
plt.plot(x,y**2, '-or')
plt.plot(x,np.cos(x),".-k")


plt.xlabel("x(in radians)")
plt.ylabel("(sin x)^2, cos x")
plt.title("Plot of trigonometric functions")

plt.grid()

plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x, np.sin(x), label="sin x")
plt.title("Graph: sin x")
plt.legend()


plt.subplot(1,2,2)
plt.scatter(x, np.sin(x)**2, label="sin2 x")
plt.title("Graph: sin^2 x")

plt.legend()
plt.show()
```

```python
for i in range(0.1,5,0.01):
    y = np.sin(x) + i
    plt.plot(x,y,label=str(i))
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure with specified figure size
plt.figure(figsize=(10, 8))  # This sets the overall figure size

# Create an axes instance with specified axis size
ax = plt.axes([0.1, 0.1, 0.85, 0.85])  # [left, bottom, width, height]

# Plot data
ax.plot(x, y)

# Set title and labels
ax.set_title('Sine Wave')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x values from 0 to pi
x = np.linspace(0, np.pi, 500)

# Plot the cosine of x
plt.plot(x, np.cos(x))  # Use x as the x-values for the plot

# Set the y-axis limits
plt.ylim(-2, 2)

# Add title and labels
plt.title("cos(x) from 0 to π")
plt.xlabel('x (radians)')
plt.ylabel('cos(x)')

# Highlight the region from index 300 to 400 in yellow
plt.axvspan(x[300], x[400], color='yellow', alpha=0.5)

# Show the plot
plt.show()
```
