```python
# Extension of the third question 
# Hrd coded data to reduce the computationa load

# returns reflectivity as a fn of N pairs
def R(N):
    n0 = 1
    ns = 1
    nh = 2.5
    nl = 1.5
    
    return ((1-(ns/n0)*(nh/nl)**(2*N))/(1+(ns/n0)*(nh/nl)**(2*N)))**2

R(10)
```

```python
from matplotlib import pyplot as plt

N = [x for x in range(1,20,1)]
R_analytical = [R(n) for n in N]

# Values obtained from the programs
R_N = [0.9995, 0.9753, 0.5982, 0.2251]
N_pairs = [10, 5, 2, 1]

plt.plot(N, R_analytical, "--k", label = "analytical", linewidth = 1.0 )
plt.plot(N_pairs, R_N, "or", label="data from simulation")
plt.legend()
plt.grid()
plt.xlabel("No of Dielectric Pairs")
plt.ylabel("Reflectivity")
```
